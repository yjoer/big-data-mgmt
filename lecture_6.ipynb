```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpec

%matplotlib inline
%config InlineBackend.figure_formats = ['retina']
```

```python
im = cv2.imread("L6/fish.jpg")
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
```

```python
fig = plt.figure(figsize=(12, 4.8))

gs = GridSpec(1, 2, width_ratios=[3, 7])

ax1 = fig.add_subplot(gs[0])
ax1.imshow(im, aspect="auto", cmap="gray")

ax2 = fig.add_subplot(gs[1])
ax2.hist(np.ravel(im), bins=256)
ax2.set_yscale("log")

plt.show()
```

```python
height, width = im.shape

mid_x = width // 2
mid_y = height // 2

top_left = im[0:mid_y, 0:mid_x]
top_right = im[0:mid_y, mid_x:width]
bottom_left = im[mid_y:height, 0:mid_x]
bottom_right = im[mid_y:height, mid_x:width]
```

```python
fig = plt.figure(figsize=(12, 9.6))

ax1 = fig.add_subplot(241)
ax1.imshow(top_left, aspect="auto", cmap="gray")

ax2 = fig.add_subplot(242, sharey=ax1)
ax2.imshow(top_right, aspect="auto", cmap="gray")

ax3 = fig.add_subplot(243, sharey=ax1)
ax3.imshow(bottom_left, aspect="auto", cmap="gray")

ax4 = fig.add_subplot(244, sharey=ax1)
ax4.imshow(bottom_right, aspect="auto", cmap="gray")

ax5 = fig.add_subplot(245)
ax5.hist(np.ravel(top_left), bins=256)
ax5.set_yscale("log")

ax6 = fig.add_subplot(246, sharex=ax5)
ax6.hist(np.ravel(top_right), bins=256)
ax6.set_yscale("log")

ax7 = fig.add_subplot(247, sharex=ax5)
ax7.hist(np.ravel(bottom_left), bins=256)
ax7.set_yscale("log")

ax8 = fig.add_subplot(248, sharex=ax5)
ax8.hist(np.ravel(bottom_right), bins=256)
ax8.set_yscale("log")

plt.show()
```
